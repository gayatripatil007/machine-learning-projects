```python
"""Demonstrates how to make a simple call to the Natural Language API."""

from google.cloud import language_v1



def complete_text_sentiment(annotations):
    score = annotations.document_sentiment.score
    magnitude = annotations.document_sentiment.magnitude
    
    print(
        "Overall Sentiment: score of {} with magnitude of {}".format(score, magnitude)
    )
    if score>=0.25:
        sentiment = "Positive"
    elif score<=(-0.25) and score>=(-1.0):
        sentiment = "Negative"
    elif score>(-0.25) and score<=(0.25):
        sentiment = "Neutral"
    
    return score, magnitude, sentiment


def sentence_level_sentiment(annotations):
    sentences_sentiments = []
    for index, sentence in enumerate(annotations.sentences):
        print(sentence.text.content)
        sentence_sentiment_score = sentence.sentiment.score
        sentence_sentiment_magnitude = sentence.sentiment.magnitude
        
        print(
            "Sentence: {} \nhas a sentiment score: {} with magnitude of {}".format(sentence.text.content, sentence_sentiment_score, sentence_sentiment_magnitude)
        )
        sentence_dict = {}
        sentence_dict['sentence'] = sentence.text.content
        sentence_dict['sentiment_score'] = sentence_sentiment_score
        sentence_dict['magnitude'] = sentence_sentiment_magnitude
        
        if sentence_sentiment_score>=0.25:
            sentence_dict['sentiment'] = "Positive"
        elif sentence_sentiment_score<=(-0.25) and sentence_sentiment_score>=(-1.0):
            sentence_dict['sentiment'] = "Negative"
        elif sentence_sentiment_score>(-0.25) and sentence_sentiment_score<=(0.25):
            sentence_dict['sentiment'] = "Neutral"
            
        sentences_sentiments.append(sentence_dict)
        
    return sentences_sentiments
        
        

    
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.kl import KLSummarizer


def get_sumy_summaries(text, sentences_count=3):

    my_parser = PlaintextParser.from_string(text, Tokenizer('english'))
    
    kl_summarizer=KLSummarizer()
    kl_summary=kl_summarizer(my_parser.document,sentences_count=sentences_count)
    
    kl_summary_output = ''
    for sentence in kl_summary:
        kl_summary_output += str(sentence) +"\n"
    return kl_summary_output


```
